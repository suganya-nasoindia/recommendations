```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
df = pd.read_csv(r"C:\Users\Lenovo\Downloads\movie_dataset.csv")
```

```python
df
```

```python
features = ['keywords', 'cast', 'genres', 'director']
for feature in features:
    df[feature] = df[feature].fillna('')
```

```python
df.info()
```

```python
def combined_features(row):
    return row['keywords']+" "+row['cast']+" "+row['genres']+" "+row['director']
df["combined_features"] = df.apply(combined_features, axis =1)
```

```python
cv = CountVectorizer()
count_matrix = cv.fit_transform(df['combined_features'])
print("Count Matrix: ",count_matrix.toarray())
```

```python
print(df["combined_features"])
```

```python
cosine_sim = cosine_similarity(count_matrix)
```

```python
print(cosine_sim)
```

```python
movie_user_like = "House of Sand and Fog"

def get_index_from(title):
    return df[df.title == title]["index"].values[0]

movie_index = get_index_from(movie_user_like)
```

```python
print(movie_index)
```

```python
similar_movies = list(enumerate(cosine_sim[movie_index]))
```

```python
print(similar_movies)
```

```python
sorted_similar_movies = sorted(similar_movies, key = lambda x:x[1], reverse = True)
print(sorted_similar_movies)
```

```python
def get_title_from_index(index):
    return df[df.index == index]["title"].values[0]

i=0
for movies in sorted_similar_movies:
    print(get_title_from_index(movies[0]))
    i = i+1;
    if i>15:
        break
```
